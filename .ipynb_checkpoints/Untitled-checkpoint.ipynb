```python
import numpy as np
import pandas as pd
xls = pd.ExcelFile("traning_source/20160419001_2016419_114348.xls")
sheetX = xls.parse(0)
print(type(sheetX))
sheetX.columns[0]
```

```python
columns1 = sheetX.loc[range(7499),sheetX.columns[1]]
columns1
```

```python
fcolumns1 = np.fft.fft(columns1)
```

```python
yf=abs(fft(fcolumns1))
yf1=abs(fft(fcolumns1))/7500
yf2 = yf1[range(int(7500/2))]
yf1
```

```python
yf2
len(yf2)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 1)
ax[0].plot(range(7499),fcolumns1)
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')

ax[1].plot(range(len(yf2)),yf2,'r') # plotting the spectrum
ax[1].set_xlabel('Freq (Hz)')
ax[1].set_ylabel('|Y(freq)|')
plt.show()
```

```python
yfcolumns1_split = np.array_split(yf2, 5, axis=0)
yfcolumns1_split0 = yfcolumns1_split[0]
yfcolumns1_split1 = yfcolumns1_split[1]
yfcolumns1_split2 = yfcolumns1_split[2]
yfcolumns1_split3 = yfcolumns1_split[3]
yfcolumns1_split4 = yfcolumns1_split[4]
fig, ax = plt.subplots(5, 1)

ax[0].plot(range(len(yfcolumns1_split0)),yfcolumns1_split0) # plotting the spectrum
ax[0].set_xlabel('Freq (Hz)')
ax[0].set_ylabel('|Y(freq)|')

ax[1].plot(range(len(yfcolumns1_split1)),yfcolumns1_split1) # plotting the spectrum
ax[1].set_xlabel('Freq (Hz)')
ax[1].set_ylabel('|Y(freq)|')

ax[2].plot(range(len(yfcolumns1_split2)),yfcolumns1_split2) # plotting the spectrum
ax[2].set_xlabel('Freq (Hz)')
ax[2].set_ylabel('|Y(freq)|')

ax[3].plot(range(len(yfcolumns1_split3)),yfcolumns1_split3) # plotting the spectrum
ax[3].set_xlabel('Freq (Hz)')
ax[3].set_ylabel('|Y(freq)|')

ax[4].plot(range(len(yfcolumns1_split4)),yfcolumns1_split4) # plotting the spectrum
ax[4].set_xlabel('Freq (Hz)')
ax[4].set_ylabel('|Y(freq)|')

plt.show()
```

```python
print("max: "+str(np.amax(yfcolumns1_split0))+" min: "+str(np.amin(yfcolumns1_split0))+" average: "+str(np.average(yfcolumns1_split0)))
print("max: "+str(np.amax(yfcolumns1_split1))+" min: "+str(np.amin(yfcolumns1_split1))+" average: "+str(np.average(yfcolumns1_split1)))
print("max: "+str(np.amax(yfcolumns1_split2))+" min: "+str(np.amin(yfcolumns1_split2))+" average: "+str(np.average(yfcolumns1_split2)))
print("max: "+str(np.amax(yfcolumns1_split3))+" min: "+str(np.amin(yfcolumns1_split3))+" average: "+str(np.average(yfcolumns1_split3)))
print("max: "+str(np.amax(yfcolumns1_split4))+" min: "+str(np.amin(yfcolumns1_split4))+" average: "+str(np.average(yfcolumns1_split4)))
```
